```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('1400-04-16.xlsx')
```

```python
df.head()
```

```python
qm_columns = [
                'Symbol', 
                'Price', 
                '1Y return', 
                '1Y return Percentile',
                '6M return', 
                '6M return Percentile',
                '3M return', 
                '3M return Percentile',
                '1M return', 
                '1M return Percentile',
                'QM Score'
                ]
qm_df = pd.DataFrame(columns = qm_columns)

for i in df.index:
        qm_df = qm_df.append(
                                pd.Series([
                                           df['symbol'].loc[i], 
                                           df['Price'].loc[i],
                                           df['1Y return'].loc[i],
                                           'N/A',
                                           df['6M return'].loc[i],
                                           'N/A',
                                           df['3M return'].loc[i],
                                           'N/A',
                                           df['1M return'].loc[i],
                                           'N/A',
                                           'N/A'
                                           ], 
                                              index = qm_columns), 
                                    ignore_index = True)
qm_df.head()
```

```python
from scipy.stats import percentileofscore
```

```python
for i in qm_df.index:
    qm_df['1Y return Percentile'].loc[i] = percentileofscore(qm_df['1Y return'],qm_df['1Y return'].loc[i])
    qm_df['6M return Percentile'].loc[i] = percentileofscore(qm_df['6M return'],qm_df['6M return'].loc[i])
    qm_df['3M return Percentile'].loc[i] = percentileofscore(qm_df['3M return'],qm_df['3M return'].loc[i])
    qm_df['1M return Percentile'].loc[i] = percentileofscore(qm_df['1M return'],qm_df['1M return'].loc[i])
    
```

```python
qm_df.head()
```

```python
from statistics import mean
for i in qm_df.index:
    qm_df['QM Score'].loc[i] = round(mean([qm_df['1Y return Percentile'].loc[i] ,
                                       qm_df['6M return Percentile'].loc[i] ,
                                       qm_df['3M return Percentile'].loc[i] ,
                                       qm_df['1M return Percentile'].loc[i]]
                                      ),1)
```

```python
qm_df.sort_values('QM Score',ascending=False,inplace = True)
qm_df.reset_index(drop=True,inplace = True)
```
